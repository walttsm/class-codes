```

import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import requests
from io import BytesIO

response = requests.get('http://www.ee.cityu.edu.hk/~lmpo/lenna/len_std.jpg')
img = Image.open(BytesIO(response.content))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```
imagem_color_arr = np.array(img)

img_gray_arr = np.mean(imagem_color_arr, axis=2)

img_gray = Image.fromarray(img_gray_arr)

plt.imshow(img_gray)
plt.axis('off')
plt.show()
```

```
negative_arr = 255 - img_gray_arr

negative = Image.fromarray(negative_arr)

plt.imshow(negative)
plt.axis('off')
plt.show()
```

```
img = np.copy(img_gray_arr)
# c = 255 / np.log(1 + np.max(img_gray_arr)) 
c = 15
log_image_arr = c * (np.log(img_gray_arr + 1))
  
# log_image = np.array(log_image, dtype = np.uint8) 
log_image = Image.fromarray(log_image_arr)

plt.imshow(log_image)
plt.title(str(c))
plt.axis('off')
plt.show() 
```

```
img = np.copy(img_gray_arr)
c = 155

gamma = 0.5

gamma_corrected_arr = np.array(c * (img / 255) ** gamma) 

gamma_image = Image.fromarray(gamma_corrected_arr)

plt.imshow(gamma_image)
plt.title(str(c))
plt.axis('off')
plt.show() 
```

```
img = np.copy(img_gray_arr)

img[img < 100]  = 0
img[(100 >= img) & (img <= 160)] = 255
img[img > 160] = 0

image = Image.fromarray(img)

plt.imshow(image)
plt.axis('off')
plt.show() 
```
